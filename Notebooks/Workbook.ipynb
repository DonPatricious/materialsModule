```python
import pandas as pd
import chardet
```

```python
# Automatically detect the encoding
with open("../Source/sampleMaterials.csv", "rb") as f:
    result = chardet.detect(f.read())
    result = result["encoding"]

print(result)
```

```python
#Loading the sample dataframe
df_materials = pd.read_csv('../Source/sampleMaterials.csv', encoding=result)
df_materials
```
